```python
### Import libraries ----
import pandas as pd
import matplotlib.pyplot as plt
```

```python
### Load data ----
df_ucdp = pd.read_csv('ucdp_month.csv')
df_ucdp = df_ucdp.set_index("date")
df_ucdp.head(5)
```

```python
### How many data points ---
df_ucdp.index
```

```python
### How many countries ----
len(list(df_ucdp.columns))
```

```python
### Check countries with high low fatalities -----
pd.DataFrame(df_ucdp.sum(axis=0)).sort_values(by=0).head(10)
```

```python
### Make summary plot -- High fatality countries -----
fig = plt.figure(figsize=(10,8))
plt.plot(df_ucdp["DR Congo (Zaire)"], color="black", linestyle="solid", label="DRC")
plt.plot(df_ucdp["Syria"], color="green", linestyle="solid", label="Syria")
plt.plot(df_ucdp["Afghanistan"], color="blue", linestyle="solid", label="Afghanistan")
plt.plot(df_ucdp["Ethiopia"], color="purple", linestyle="solid", label="Ethiopia")
plt.plot(df_ucdp["Iraq"], color="orange", linestyle="solid", label="Iraq")

# Remove xticks
# Source: https://stackoverflow.com/questions/12998430/remove-xticks-in-a-matplotlib-plot
plt.xticks([], [])
plt.yticks(size=12)
plt.legend()

# Save
plt.savefig(f"summary_high.png",
            bbox_inches="tight",
            dpi=200)
```

```python
### Load predictions  -----
preds_lasso = pd.read_csv('preds_lasso.csv')

# Source: https://stackoverflow.com/questions/34049618/how-to-add-a-suffix-or-prefix-to-each-column-name
preds_lasso = preds_lasso.add_suffix('_lasso')
preds_lasso.head(3)
```

```python
### Load predictions  -----
preds_rf = pd.read_csv('preds_rf.csv')

# Source: https://stackoverflow.com/questions/34049618/how-to-add-a-suffix-or-prefix-to-each-column-name
preds_rf = preds_rf.add_suffix('_rf')
preds_rf.head(3)
```

```python
### Load predictions  -----
preds_nn = pd.read_csv('preds_nn.csv')

# Source: https://stackoverflow.com/questions/34049618/how-to-add-a-suffix-or-prefix-to-each-column-name
preds_nn = preds_nn.add_suffix('_nn')
preds_nn.head(3)
```

```python
### Load test data  -----
df_test = pd.read_csv('df_test.csv')

# Source: https://stackoverflow.com/questions/34049618/how-to-add-a-suffix-or-prefix-to-each-column-name
df_test = df_test.add_suffix('_actuals')
df_test.head(3)
```

```python
### Merge ----
preds_all = df_test.join(preds_lasso)
preds_all = preds_all.join(preds_rf)
preds_all = preds_all.join(preds_nn)
preds_all.set_index(df_test.index)
```

```python
### Prediction plot ------
fig = plt.figure(figsize=(10,8))
plt.plot(preds_all["date_actuals"], preds_all["Afghanistan_actuals"], color="black", linestyle="solid", label="Actuals")
plt.plot(preds_all["Afghanistan_rf"], color="green", linestyle="solid", label="Random Forest")
plt.plot(preds_all["Afghanistan_lasso"], color="blue", linestyle="solid", label="LASSO")
plt.plot(preds_all["Afghanistan_nn"], color="purple", linestyle="solid", label="Neural Net")

plt.xticks(rotation=30, size=5, ha='right')
plt.legend()
plt.title("Actuals vs Perdictions for Afghanistan")
# Save
plt.savefig(f"preds_Afghanistan.png",
            bbox_inches="tight",
            dpi=200)
```

```python
### Prediction plot ------
fig = plt.figure(figsize=(10,8))
plt.plot(preds_all["date_actuals"], preds_all["Sudan_actuals"], color="black", linestyle="solid", label="Actuals")
plt.plot(preds_all["Sudan_rf"], color="green", linestyle="solid", label="Random Forest")
plt.plot(preds_all["Sudan_lasso"], color="blue", linestyle="solid", label="LASSO")
plt.plot(preds_all["Sudan_nn"], color="purple", linestyle="solid", label="Neural Net")

plt.xticks(rotation=30, size=5, ha='right')
plt.legend()
plt.title("Actuals vs Perdictions for Sudan")
# Save
plt.savefig(f"preds_Sudan.png",
            bbox_inches="tight",
            dpi=200)
```
