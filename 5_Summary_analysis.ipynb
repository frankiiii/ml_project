```python
### Load predictions LASSO -----
preds_lasso = pd.read_csv('preds_lasso.csv')

# Add suffix
# Source: https://stackoverflow.com/questions/34049618/how-to-add-a-suffix-or-prefix-to-each-column-name
preds_lasso = preds_lasso.add_suffix('_lasso')
preds_lasso.head(3)
```

```python
### Load predictions random forest -----
preds_rf = pd.read_csv('preds_rf.csv')

# Add suffix
# Source: https://stackoverflow.com/questions/34049618/how-to-add-a-suffix-or-prefix-to-each-column-name
preds_rf = preds_rf.add_suffix('_rf')
preds_rf.head(3)
```

```python
### Load predictions neural net -----
preds_nn = pd.read_csv('preds_nn.csv')

# Add suffix
# Source: https://stackoverflow.com/questions/34049618/how-to-add-a-suffix-or-prefix-to-each-column-name
preds_nn = preds_nn.add_suffix('_nn')
preds_nn.head(3)
```

```python
### Load test data -----
df_test = pd.read_csv('df_test.csv')

# Add suffix
# Source: https://stackoverflow.com/questions/34049618/how-to-add-a-suffix-or-prefix-to-each-column-name
df_test = df_test.add_suffix('_actuals')
df_test.head(3)
```

```python
### Merge predictions and actuals ----
preds_all = df_test.join(preds_lasso)
preds_all = preds_all.join(preds_rf)
preds_all = preds_all.join(preds_nn)
preds_all.head(10)
```

```python
### Prediction plot for Afghanistan ------

# Initiate plot
fig = plt.figure(figsize=(10,8))

# Plot 
plt.plot(preds_all["date_actuals"], 
         preds_all["Afghanistan_actuals"], 
         color="black", 
         linestyle="solid", 
         label="Actuals")

plt.plot(preds_all["Afghanistan_lasso"], 
         color="blue", 
         linestyle="solid", 
         label="LASSO")

plt.plot(preds_all["Afghanistan_rf"], 
         color="green", 
         linestyle="solid", 
         label="Random Forest")

plt.plot(preds_all["Afghanistan_nn"], 
         color="purple", 
         linestyle="solid", 
         label="Neural Net")


# xticks
plt.xticks(rotation=30, size=5, ha='right')

# Legend
plt.legend()

# Title
plt.title("Actuals vs Perdictions for Afghanistan")

# Save
plt.savefig(f"preds_Afghanistan.png",
            bbox_inches="tight",
            dpi=200)
```

```python
### Prediction plot for Sudan ------

# Initiate plot
fig = plt.figure(figsize=(10,8))

# Plot
plt.plot(preds_all["date_actuals"], 
         preds_all["Sudan_actuals"], 
         color="black", 
         linestyle="solid", 
         label="Actuals")

plt.plot(preds_all["Sudan_lasso"], 
         color="blue", 
         linestyle="solid", 
         label="LASSO")

plt.plot(preds_all["Sudan_rf"], 
         color="green", 
         linestyle="solid", 
         label="Random Forest")

plt.plot(preds_all["Sudan_nn"], 
         color="purple", 
         linestyle="solid", 
         label="Neural Net")

# xticks
plt.xticks(rotation=30, size=5, ha='right')

# Legend
plt.legend()

# Title
plt.title("Actuals vs Perdictions for Sudan")

# Save
plt.savefig(f"preds_Sudan.png",
            bbox_inches="tight",
            dpi=200)
```
